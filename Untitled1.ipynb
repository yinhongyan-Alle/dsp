```python
import numpy as np

hs = np.fft.rfft(wave.ys)#使用快速傅里叶变换（FFT）计算信号的频谱
hs
```

```python
n = len(wave.ys)                 # 获取信号样本的数量
d = 1 / wave.framerate           # 计算样本之间的时间间隔
fs = np.fft.rfftfreq(n, d)#计算FFT的频率轴
fs
```

```python
import matplotlib.pyplot as plt

magnitude = np.absolute(hs)
plt.plot(fs, magnitude)
decorate(xlabel='Frequency (Hz)')
```

```python
import matplotlib.pyplot as plt

magnitude = np.absolute(hs)#计算频谱的幅值
plt.plot(fs, magnitude)
decorate(xlabel='Frequency (Hz)')
```

```python
angle = np.angle(hs)#计算频谱的相位
plt.plot(fs, angle)
decorate(xlabel='Phase (radian)')
```

```python
import random
random.shuffle(angle)#：将相位数组随机打乱
plt.plot(fs, angle)#绘制打乱后的相位图
decorate(xlabel='Phase (radian)')
```

```python
i = complex(0, 1)#创建一个虚数单位。
spectrum = wave.make_spectrum()#计算信号的频谱。
spectrum.hs = magnitude * np.exp(i * angle)#根据幅值和相位重建复数频谱
```

```python
wave2 = spectrum.make_wave()#根据重建的频谱生成新的信号波形
wave2.normalize()#对新信号波形进行归一化处理
segment = wave2.segment(duration=0.005)#从新信号波形中截取一段持续时间为0.005秒的片段。
segment.plot()
decorate(xlabel='Time (s)')
```

```python
def view_harmonics(freq, framerate):#定义一个函数，用于绘制锯齿波的频谱
    """Plot the spectrum of a sawtooth signal.
    
    freq: frequency in Hz
    framerate: in frames/second
    """
    signal = SawtoothSignal(freq)#创建一个频率为freq的锯齿波信号对象
    wave = signal.make_wave(duration=0.5, framerate=framerate)#生成信号的波形数据。
    spectrum = wave.make_spectrum()#计算信号的频谱
    spectrum.plot(color='C0')#绘制信号的频域图
    decorate(xlabel='Frequency (Hz)', ylabel='Amplitude')
    display(wave.make_audio())#显示音频文件
```

```python
from ipywidgets import interact, interactive, fixed#导入Jupyter Notebook的交互式小部件库。
import ipywidgets as widgets

slider1 = widgets.FloatSlider(min=100, max=10000, value=100, step=100)
slider2 = widgets.FloatSlider(min=5000, max=40000, value=10000, step=1000)#创建一个滑动条，用于调整频率。
interact(view_harmonics, freq=slider1, framerate=slider2);#slider2 = widgets.FloatSlider(min=5000, max=40000, value=10000, step=1000)：创建另一个滑动条，用于调整
```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from thinkdsp import SawtoothSignal, decorate
from IPython.display import display, Audio
from ipywidgets import interact, FloatSlider
import ipywidgets as widgets

# 定义一个函数，用于绘制锯齿波的频谱
def view_harmonics(freq, framerate):
    signal = SawtoothSignal(freq)  # 创建一个频率为freq的锯齿波信号对象
    wave = signal.make_wave(duration=0.5, framerate=framerate)  # 生成信号的波形数据
    spectrum = wave.make_spectrum()  # 计算信号的频谱
    spectrum.plot(color='C0')  # 绘制信号的频域图
    decorate(xlabel='Frequency (Hz)', ylabel='Amplitude')  # 添加图表标签
    display(wave.make_audio())  # 显示音频播放控件

# 创建滑块，用于调整频率和帧率
slider1 = widgets.FloatSlider(
    min=100,
    max=10000,
    value=1000,
    step=100,
    description='Frequency (Hz):',
    continuous_update=False
)
slider2 = widgets.FloatSlider(
    min=5000,
    max=40000,
    value=10000,
    step=1000,
    description='Frame rate (Hz):',
    continuous_update=False
)

# 创建交互式界面
interact(view_harmonics, freq=slider1, framerate=slider2)
```
